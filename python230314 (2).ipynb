## 사칙연산
- 정수형 사칙연산

```python
a = 6
b = 2

print(a+b)   #덧셈
print(a-b)   #뺄셈
print(a*b)   #곱셈
print(a/b)   #나누기 - 실수형으로 나온다
print(a//b)  #몫
print(a**b)  #제곱
print(a%b)   #나머지
```

### 실수형 사칙연산

```python
a = 6.0
b = 2.0

print(a+b)   #덧셈
print(a-b)   #뺄셈
print(a*b)   #곱셈
print(a/b)   #나누기 - 실수형으로 나온다
print(a//b)  #몫
print(a**b)  #제곱
print(a%b)   #나머지
```

### 논리형 연산자

### 슬라이싱

```python
greeting = "Hello World!"
print(greeting[0:5])  #[start_index:end_index:step_num]---> end_index :  마지막 인덱스 + 
print(greeting[:])
print(greeting[6:])
print(greeting[3:8])
print(greeting[5:-1])
print(greeting[0:9:2])
```

### 문자열의 함수들

### 중첩리스트(Nested List)
- 리스트내의 또 다른 리스트

```python
a = [['apple','banana','cherry'],1]
print(a)
print(a[0][2][2])
# cherry 글자에서 알파벳 e만 가져오세요.
```

```python
a = [['apple','banana','cherry'],1]
# cherry 글자에서 알파벳 e만 가져오세요.
temp = a[0]
temp2 = temp[2]
temp2[2]
```
