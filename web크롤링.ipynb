```python
import requests
from bs4 import BeautifulSoup

url = 'http://media.daum.net/digital/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

articles = soup.find_all('a', class_='link_txt')
for article in articles:
    print(article.text)
```
