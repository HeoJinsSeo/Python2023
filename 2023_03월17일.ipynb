## 데이터 핸들링
- 행과 열을 다루는 방법

```python
lemonade.head()
```

- 새로운 칼럼 추가


```python
lemonade['Sold'] = lemonade['Lemon'] + lemonade['Orange']
lemonade.head(3)
```

- Revenue 칼럼 : 매출 계산

```python
lemonade['Revenue'] = lemonade['Price'] * lemonade['Sold']
lemonade.head(3)
```

## 특정 컬럼 삭제

```python
# axis = 1, 컬럼을 삭제하겠다!!
lemonade_col_drop = lemonade.drop('Sold', axis = 1)  # axis = 1 :columns (열)
lemonade_col_drop.head(3)                            # axis = 0 : 행
```

```python
# axis = 0, 조건에 맞는 행을 삭제하겠다. 
lemonade_row_drop = lemonade.drop(0, axis = 0)
lemonade_row_drop.head()
```

- 행을 삭제한 후, 인덱스 번호를 초기화
 + 0번째부터 순차적으로 시작하도록 변환

```python
lemonade_row_drop = lemonade.drop(3, axis = 0)
lemonade_row_drop.reset_index(drop=True)    # --->행을 초기화하는 명령어
```

## Group by
- SQL Groupby 기능적으로 동일
--> Location 컬럼만 사용

```python
lemonade.groupby(by='Location').count().T  #.T는 Transform 행과 열 바꾸기
```

### 집계함수 (Aggregation)
- 그룹바이 및 집계 ---> 피벗테이블(피벗: 자료를 요약함)

```python
import numpy as np
lemonade.groupby('Location')['Revenue'].agg([max, min, np.mean])
```

```python
import numpy as np
lemonade.groupby('Location')[['Revenue', 'Temperature']].agg([max, min, np.mean, np.std, sum])
```
